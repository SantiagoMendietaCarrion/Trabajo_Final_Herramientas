```python
import pandas as pd
import pymysql
import sqlalchemy as db
```

```python
from faker import Faker
fake = Faker()
```

```python
lista = []
```

```python
for x in range(0,1000):
    lista.append({'Nombre': fake.first_name(),'Apellido':fake.last_name(),'Dirección':fake.address(),'Correo electrónico':fake.email(),'Fecha de nacimiento':fake.date_of_birth()})
```

```python
lista
```

```python
data = pd.DataFrame(lista)
```

```python
rango=range(1,len(data)+1)
data.index=rango
```

```python
data
```

```python
cedula=1102792826
def generar_cedula(d):
    global cedula
    cedula+=1
    return cedula
```

```python
data['Cédula']=data['Nombre'].apply(generar_cedula)
```

```python
data = data[['Cédula'] + [col for col in data.columns if col != 'Cédula']]
```
